### Дропнуть NA, провести STL анализ, который ищет во временном ряде тренд, ищет сезонные периоды посещений и остаток от того, что остаётся

```python
df_year.drop(["na"], axis=1, inplace=True)

df_year
```

```python
df_year.to_csv('Визиты_за_год.csv')
```

```python
import statsmodels.api as sm
```

```python
# Выполняем разложение STL(период=месяцу(30 дням) model='additive'- компоненты складываются
stl_decomposition = sm.tsa.seasonal_decompose(df_year, model='additive', period = 30)
```

```python
# Визуализация разложения
fig, ax = plt.subplots(4, 1, figsize=(10, 8))
 
ax[0].plot(df_year, label='Исходный ряд.Визиты')
ax[0].legend()
 
ax[1].plot(stl_decomposition.trend, label='Тренд')
ax[1].legend()
 
ax[2].plot(stl_decomposition.seasonal, label='Сезонность')
ax[2].legend()
 
ax[3].plot(stl_decomposition.resid, label='Остатки')
ax[3].legend()
 
plt.tight_layout()
plt.show()
```

## ВЫВОД:
### По "Тренду" видно увеличение активности зимой и весной, падение в летний период, в "Сезонности" (представлен ввиде повторяющегося графика) видна месячная (т.к. период выставлен месяц) и недельная (внутри месяца три пика) сезонность

```python
# Выполняем разложение STL, model='multiplicative'- компоненты перемножаются
stl_decomposition = sm.tsa.seasonal_decompose(df_year, model='multiplicative', period = 30)
```

```python
# Визуализация разложения
fig, ax = plt.subplots(4, 1, figsize=(10, 8))
 
ax[0].plot(df_year, label='Исходный ряд.Визиты')
ax[0].legend()
 
ax[1].plot(stl_decomposition.trend, label='Тренд')
ax[1].legend()
 
ax[2].plot(stl_decomposition.seasonal, label='Сезонность')
ax[2].legend()
 
ax[3].plot(stl_decomposition.resid, label='Остатки')
ax[3].legend()
 
plt.tight_layout()
plt.show()
```

## ВЫВОД
#### Проанализировали остаток, с точи зренеия того, что он ожидаемо нормально  распределён, а хвосты- аномалии, хорошо видны на бокс плоте, усы-это разброс. БоксПлот показывает разброс, от"0", размеры коробки-от медианы-это примерный интервал остатка ("ошибки" декомпозиции), а точки сверху-это выбросы, то есть особый случай, который можно рассмотреть, можно их потом выделить и посмотреть, куда они попадают, на графике их найти, все ли они попадают на один период или они распределны в какие-то, то есть какое-то событие было, то есть была аномали-большое количество поситителей, скачки посещаемости, интересно посмотреть когда это было и подумать, что в это время было-рекламная компания или что-то ещё.
